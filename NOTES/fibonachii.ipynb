```python
#fibonachii
#write a python programm to  print factorial of n number (user input)
n = int(input("Enter a number:"))
fact=1 
for i in range(1,n+1):
    fact=fact*i
print(fact)
```

```python
#write a pp to print nth element of fibonacci
a=0
b=1
n = int(input("Enter number: "))
for i in range(2,n):
    c=a+b
    a=b
    b=c
print(c) 
```
