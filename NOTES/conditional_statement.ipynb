```python
#questions
# Taking input for three subjects
science = int(input("Enter Science marks: "))
lang = int(input("Enter Language marks: "))
math = int(input("Enter Math marks: "))


obtained_marks = science + lang + math
print("Obtained Marks:", obtained_marks)


per = (obtained_marks / 300) * 100
print("Percentage:", per, "%")


if per > 65:
    print("1st Division")
elif per > 50:
    print("2nd Division")
elif per >= 33:
    print("3rd Division")
else:
    print("Fail")
    
           
```

```python
# Program to check if a year is a leap year

year = int(input("Enter any year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
n = int(input("Enter a number:"))

sum_even = 0
sum_odd = 0

for i in range(1, n+1):
    sum_even += 2*i       
    sum_odd += 2*i - 1   

print("Sum of first", n, "even numbers =", sum_even)
print("Sum of first", n, "odd numbers  =", sum_odd)
```
