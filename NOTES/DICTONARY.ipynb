```python
marks = {
    "Ankit": 90,
    "Ravi": 85,
    "Neha": 90,
    "Aman": 80
}

grouped = {}

for i, j in marks.items():
    if j not in grouped:
        grouped[j] = []
    grouped[j].append(i)

print(grouped)
```

```python
sales = [ {"product": "Laptop", "amount": 50000}, {"product": "Mobile", "amount": 20000}, {"product": "Laptop", "amount": 55000},{"product": "Tablet", "amount": 15000},
    {"product": "Mobile", "amount": 25000}]

total_sales = {}

for item in sales:
    product = item["product"]
    amount = item["amount"]

    if product in total_sales:
        total_sales[product] += amount
    else:
        total_sales[product] = amount

print(total_sales)                  
```

```python
scores = {
    "Math": 78,
    "Python": 92,
    "SQL": 85,
    "PowerBI": 70
}

min_val = min(scores.values())
max_val = max(scores.values())

normalized = {}

for subject, score in scores.items():
    normalized[subject] = round((score - min_val) / (max_val - min_val), 2)

print(normalized)
```

```python
docs = {
    1: "Python is great and Python is fast",
    2: "Data science uses Python",
    3: "Python and SQL are powerful"
}

index = {}

for doc_id, text in docs.items():
    words = text.lower().split()
    for word in words:
        if len(word) >= 3:
            if word not in index:
                index[word] = []
            if doc_id not in index[word]:
                index[word].append(doc_id)

print(index)
```

```python
nums = [10, 20, 30, 40, 50]

result = {}

for i in range(len(nums)):
    start = max(0, i - 2)
    window = nums[start:i + 1]
    result[i] = sum(window) / len(window)

print(result)
```
