```python
#direct iteration
l=[10,20,30,"hello",1000,10.5,11] 
for i in l:
    print(i)
```

```python
#indirect iteration
l=[10,20,30,"hello",1000,10.5,11] 
for i in range(len(l)):
    print(l[i])
```
